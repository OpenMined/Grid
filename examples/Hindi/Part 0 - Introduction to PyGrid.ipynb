# PyGrid क्या है?


PyGrid (जिसे Grid के रूप में भी जाना जाता है) डेटा मालिकों और डेटा वैज्ञानिकों के लिए एक सहयोगी नेटवर्क प्लेटफ़ॉर्म है जो PySyft लाइब्रेरी का उपयोग करके AI मॉडल को सामूहिक रूप से प्रशिक्षित और साझा कर सकता है, जो PyTorch और Tensorflow जैसी Deep Learning लाइब्रेरी का विस्तार करता है।

### परिचय


PyGrid प्लेटफ़ॉर्म का उद्देश्य मॉडल को प्रशिक्षित करने, प्रबंधित करने और साझा करने के लिए उपयोग किया जाने वाला सुरक्षित सहकर्मी होना है। एक Grid श्रमिकों का एक संग्रह है जो पारदर्शी तरीके से संदेशों और tensors का आदान-प्रदान कर सकता है।


Grid वर्तमान में समर्थन करता है:

* आप जैसे दूरदराज के श्रमिकों के बीच संदेशों का आदान-प्रदान आमतौर पर वर्चुअल वर्कर्स के साथ PySyft का उपयोग करना होगा
* Machine Learning एक सेवा के रूप में (MLaaS): ग्रिड का उपयोग करके मॉडल की मेजबानी और query करना
* एक सेवा (ईएमएलएएस) के रूप में Encrypted Machine Learning: ग्रिड का उपयोग करके Encrypted किए गए मॉडल की मेजबानी और query करना
* Heroku पर श्रमिकों को तैनात करने और चलाने के लिए समर्थन

Grid जल्द ही समर्थन करेगा:

* वितरित श्रमिकों को सीधे प्रबंधित करने की आवश्यकता के साथ Federated Learning प्रक्रियाएं।

### उद्देश्य

- ***सुरक्षित संघबद्ध शिक्षण मंच बनाना***
  - बौद्धिक संपदा (IP) को सुरक्षित रखें
  - डेटा गोपनीयता को सुरक्षित रखें
  - मॉडल का दुरुपयोग कम करें
- ***एक सहयोगी संघबद्ध शिक्षण मंच बनाना***
  - उपयोगकर्ताओं को सादे पाठ / अज्ञात डेटासेट भेजने की अनुमति देने के लिए जिसका उपयोग Grid प्लेटफ़ॉर्म पर अन्य उपयोगकर्ताओं द्वारा किया जा सकता है
  - उपयोगकर्ताओं को ग्रिड प्लेटफॉर्म पर अन्य उपयोगकर्ताओं द्वारा उपयोग किए जाने वाले / सुधार किए जाने वाले सादे पाठ / अज्ञात मॉडल भेजने की अनुमति देने के लिए
  - ग्रिड प्लेटफॉर्म पर उपलब्ध डेटासेट का उपयोग करते हुए फेडरेटेड लर्निंग का प्रदर्शन करना
- ***मॉडल में सुरक्षित तरीके से योगदान***
  - Grid नेटवर्क पर होस्ट किए गए मॉडल पर रिमोट प्लेन टेक्स्ट / एनोनिमस इनफॉर्म्स चलाएं
  - Grid नेटवर्क पर होस्ट किए गए मॉडल डाउनलोड करें
  - अनाम डेटासेट का उपयोग करके मनमाना अंकगणितीय / तर्क कंप्यूटिंग करें
  - गलती सहिष्णुता प्रदान करें
  
  


### कार्यान्वयन

PyGrid रिपॉजिटरी में मुख्य रूप से 2 फ़ोल्डर होते हैं:

#### एप्लिकेशन

`एप्लिकेशन` मूल रूप से REDIS डेटाबेस के साथ नियमित फ्लास्क सर्वर हैं। हम इन सर्वर को Grid वर्कर्स से बात करने की क्षमता, डेटाबेस में tensor और मॉडल स्टोर करने की सुविधा देते हैं ताकि फॉल्ट टॉलरेंस सुनिश्चित किया जा सके।

यह विचार है कि Grid वर्कर्स एक ऐप के साथ संवाद करने का एक तरीका है।

#### गेटवे

`गेटवे` एक विशेष DNS घटक की तरह काम करता है लेकिन यह डोमेन नाम के बजाय प्रश्नों द्वारा नोड्स को रूट करेगा।
Grid गेटवे को Grid नेटवर्क पर जुड़े सभी Grid नोड्स (जो कार्यकर्ता एक दूसरे से जुड़े हुए हैं) के पते और आईडी जानने की आवश्यकता है।


एक उपयोगकर्ता Grid गेटवे के साथ बात कर सकता है और प्रतीत होता है कि कई दूरस्थ नोड्स पर ऑपरेशन बिना ऑपरेशन के भी पता चलता है कि वे नेटवर्क पर मौजूद हैं।
